```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located
file = "Resources/crime_incident_data2017.csv"
# Import the data into a Pandas DataFrame
original_df = pd.read_csv(file, encoding="ISO-8859-1")
original_df.head()
```

```python
# look for missing values
original_df.count()
```

```python
# drop null rows
original_df = original_df.dropna(how="any")
```

```python
# verify counts
original_df.count()
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
original_df["Offense Type"].value_counts()
```

```python
# Combine similar offenses
original_df['Offense Type'] = original_df['Offense Type'].replace(
    {'Theft of Motor Vehicle Parts or Accessories': 'Motor Vehicle Theft', 'Theft From Motor Vehicle': 'Motor Vehicle Theft'})
```

```python
# Check to see if you comnbined similar offenses correctly in "Offense Type".
original_df["Offense Type"].value_counts()
```

```python
# Create a new DataFrame that looks into a specific neighborhood
sw_hills_df = original_df[original_df.loc[:,"Neighborhood"] == "Southwest Hills"]
sw_hills_df
```
