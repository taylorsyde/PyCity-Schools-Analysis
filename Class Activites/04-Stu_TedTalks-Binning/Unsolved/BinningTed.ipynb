```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/ted_talks.csv"
ted_df = pd.read_csv(csv_path)

ted_df.head()
```

```python
# Figure out the minimum and maximum views for a TED Talk
max_views = ted_df["views"].max()
min_views = ted_df["views"].min()
```

```python
# Create bins in which to place values based upon TED Talk views
bins = [0, 199999, 399999, 599999, 799999, 999999,
        1999999, 2999999, 3999999, 4999999, 50000000]

# Create labels for these bins
group_names = ["Under 200k Views", "200k to 400K Views","400k to 600K Views","600k to 800K Views","800k to 1 Million Views",
               "1-2 Million Views","2-3 Million Views","3-4 Million Views","4-5 Million Views","Over 5 million"]
```

```python
# Slice the data and place it into bins
ted_df["View Range"] = pd.cut(ted_df["views"], bins, labels=group_names, include_lowest=True)
ted_df
```
